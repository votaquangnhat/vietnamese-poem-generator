```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else 'cpu')
device
```

```python
from datasets import load_dataset

dataset = load_dataset("phamson02/vietnamese-poetry-corpus")
data = dataset['train']
print(set(data['genre']))
```

```python
selected_genre = "lục bát"
hf_repo_name = "vietnamese-poem-hustgpt2-lucbat"
#MODEL_NAME = 'danghuy1999/gpt2-viwiki'
#MODEL_NAME = 'VietAI/gpt-neo-1.3B-vietnamese-news'
MODEL_NAME = 'NlpHUST/gpt2-vietnamese'
filtered_data = data.filter(lambda x: x['genre'] == selected_genre)
set(filtered_data['genre'])
```

```python
poem_dataset = filtered_data.train_test_split(test_size=0.2, seed=42)  # 80% train, 20% test

print("Training set size:", len(poem_dataset['train']))
print("Test set size:", len(poem_dataset['test']))
poem_dataset
```

```python
poem_dataset['train'][0]
```

```python
def preprocess(batch):
    # Assuming the text column is named 'text'
    batch['content'] = [text.replace('<\n>', '\n') for text in batch['content']]
    return batch


poem_dataset = poem_dataset.map(preprocess, batched=True)
poem_dataset['train'][0]['content']
```

```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)

tokenizer.pad_token = tokenizer.eos_token

def tokenize_function(examples):
    return tokenizer(examples['content'], truncation=True, padding="max_length", max_length=1024, return_tensors='pt')

# Apply tokenization
tokenized_poem_dataset = poem_dataset.map(tokenize_function, batched=True, num_proc=4, remove_columns=poem_dataset['train'].column_names)

tokenized_poem_dataset.set_format("torch")

tokenized_poem_dataset['train'][0]
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

from transformers import GPT2LMHeadModel, TrainingArguments, Trainer

model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)
model.resize_token_embeddings(len(tokenizer))

training_args = TrainingArguments(
    output_dir=hf_repo_name,
    eval_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=1,           # Keep only the last checkpoint
    load_best_model_at_end=True,  # Load the best model after training
    learning_rate=2e-5,
    lr_scheduler_type="linear",
    warmup_steps=500,               # Warm-up steps for scheduler
    weight_decay=0.01,
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    fp16=True,                      # Mixed precision for speedup
    push_to_hub=True
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_poem_dataset['train'],
    eval_dataset=tokenized_poem_dataset['test'],
    data_collator=data_collator,
    tokenizer=tokenizer,
)

trainer.train()
```

```python
text = "con tim"
input_ids = tokenizer.encode(text, return_tensors='pt').to(device)
max_length = 300

sample_outputs = model.generate(input_ids,
                                do_sample=True,
                                max_length=max_length,
                                min_length=max_length,
                                max_new_tokens=50,
                                top_k=50,
                                top_p=0.95,
                                num_beams=5,
                                early_stopping=True,
                                no_repeat_ngram_size=2,
                                num_return_sequences=3)

output = tokenizer.decode(sample_outputs[0])

for line in output.split('\n'):
    print(line)
```
